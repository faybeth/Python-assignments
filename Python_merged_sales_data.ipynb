## Merging in Python
- Merging in python is the process of combining two or more DataFrames which are tables into one based on a common column.
- Merging is important because it allows us to bring related data from different tables into one dataset, it adds context to raw data by adding meaningful attributes to it.It also makes data ready for analysis and reporting.

## Knowing what to merge
- Identify the fact table, a fact table contains transactions or events(what actually happened) and is usually the biggest table.
Example a customerID 101 bought a product with ID 201 on 1st Jan 2023.
    - the sales table is the fact table it has transactions that is the salesID, customerID,productID 
- Identify the demension tables, a demension table describe attributes.(describes what happened)
    - Adds attributes example customerID 101 ="Alice",Region="Nairobi".productID 201 is "Maize", category="Grains"
- Dimensions:

Customers (join on CustomerID)

Products (join on ProductID)

Calendar (join SaleDate ↔ Date)

Regions (join Region id ↔ Region ID)

Categories (join on Category text)    

Analytics/reporting/Fact+ dimension → LEFT JOIN

Matching only/ for comparing or filtering datasets → INNER JOIN

Coverage checks → OUTER JOIN 
- eg when you want to check which customers existin the customers table but never appear in sales 

```python
# Merge Sales with Customers using the left join
merged = sales.merge(customers, on="CustomerID", how="left")
```

```python
# Merge with Products using the left join
merged = merged.merge(products, on="ProductID", how="left")
```

```python
# Merge with Calendar takes the saleDate column from sales and matches it with the date column in the calender table, left join because it kepps all sales 
merged = merged.merge(calendar, left_on="SaleDate", right_on="Date", how="left")
```

```python
# Merge Customers → Regions, 
# left_on uses the column with Region id in the merged dataframe which comes from customers
#right_on uses the column name Region ID in the Regions table 
# we can't use on because of id and ID 
merged = merged.merge(regions, left_on="Region id", right_on="Region ID", how="left")
```

```python
# Merge Products → Categories
merged = merged.merge(categories, left_on="Category", right_on="Category", how="left")
```
